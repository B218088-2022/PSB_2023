## Solving ordinary differential equations – simulating deterministic systems

As an example, we will consider the Michaelis-Menten model of enzyme catalysis. For substrate $S$ turning into product $P$ catalysing by enzyme $E$, which forms a complex with the substrate $C$, we have
$$
\begin{align}
\frac{dS}{dt} &= -f E S + b C \\
\frac{dE}{dt} &= -f E S + b C + k C \\
\frac{dC}{dt} &= f E S - b C - k C \\
\frac{dP}{dt} &= k C
\end{align}
$$
which we would like to solve from some starting time to a finishing time.

First, we need to import the modules needed, with <tt>odeint</tt> being a function that integrates differential equations:

```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
from scipy.integrate import odeint
```

Second, we must code up the model's differential equations into a function, which has to have a specific format for <tt>odeint</tt> to work.

```python
def mm(y, t):
    f= 1
    b= 0.01
    k= 2
    # y需要是一个含有几个数的list
    # y[0] = E, y[0] = S, y[2] = C, y[3] = P
    dydt= np.empty(len(y)) # 定义四个函数
    dydt[0]= -f*y[1]*y[0] + b*y[2] # y[0] = E, y[0] = S, y[2] = C
    dydt[1]= -f*y[1]*y[0] + b*y[2] + k*y[2]
    dydt[2]= f*y[1]*y[0] - b*y[2] - k*y[2]
    dydt[3]= k*y[2]
    return dydt
```

Check that you understand what each <tt>y</tt> is and how this code relates to the model.

```python
mm([0.1, 4, 0.3, 0], 0.1)
```

Third, we should specify the time of interest and the initial values of all the variables (concentrations here) in the model:
定义time array，初始化所有变量的初始值

```python
t= np.linspace(0, 800, 100)
yi= np.array([1000, 1, 0, 0]) #initialize the initial values of all variables(E, S, C, P)
```

Finally, we can call <tt>odeint</tt>:

```python
y= odeint(mm, yi, t)
y
```

This command generates a 2d array, <tt>y</tt>, where each column corresponds to the one of the variables in the model and each row corresponds to a time point. For example, to plot the product, we can use:

```python
plt.figure()
#for i in range(4):
plt.plot(t, y[:,3], t, y[:,0]) #代表第四列的所有元素
plt.xlabel('time')
plt.ylabel('concentration')
plt.title('P and S')
plt.legend(labels = ["P", "S"], loc = "best")
plt.show()
```

```python
fig = plt.figure(figsize= (20, 12))
plot = plt.subplot(311)
plot.plot(t, y[:,0])
plt.xlabel('time')
plt.ylabel('concentration')
plt.title('S')

plot = plt.subplot(312)
plot.plot(t, y[:, 2])
plt.xlabel('time')
plt.ylabel('concentration')
plt.title('C')

plot = plt.subplot(313)
plot.plot(t, y[:, 3])
plt.xlabel('time')
plt.ylabel('concentration')
plt.title('P')
plt.show()
```

We can re-code the Michaelis-Menten model to be easier to read:

```python
def mm(y, t):
    f= 1
    b= 0.01
    k= 2
    s, e, c, p= y
    dydt= np.empty(len(y))
    dydt[0]= -f*e*s + b*c
    dydt[1]= -f*e*s + b*c + k*c
    dydt[2]= f*e*s - b*c - k*c
    dydt[3]= k*c
    return dydt
```

We also may wish to change the parameters and so have the parameters as an input to the function:

```python
def mmp(y, t, params): #如果将params改成可变参数*params，则后面给odeint传递参数的时候需要将params定义成一个tuple
    f, b, k= params #现在f, b, k是需要传递给函数的参数
    s, e, c, p= y
    dydt= np.empty(len(y))
    dydt[0]= -f*e*s + b*c
    dydt[1]= -f*e*s + b*c + k*c
    dydt[2]= f*e*s - b*c - k*c
    dydt[3]= k*c
    return dydt
```

We can define <tt>params</tt> as, for example,

```python
params= np.array([1, 0.01, 2])
(params,)
```

Running <tt>odeint</tt> is now different because our function expects three arguments and not the default of two. We use:

```python
Param = (1, 0.01, 2)
Param
```

```python
y= odeint(mmp, yi, t, args= (params,))
#y= odeint(mmp, yi, t, args= Param)
y
```

Alternatively, you could use:

```python
def mmpa(y, t, f, b, k):
    s, e, c, p= y
    dydt= np.empty(len(y))
    dydt[0]= -f*e*s + b*c
    dydt[1]= -f*e*s + b*c + k*c
    dydt[2]= f*e*s - b*c - k*c
    dydt[3]= k*c
    return dydt
```

where <tt>params</tt> must be defined as a tuple – with normal brackets not square brackets:

```python
params= (1, 0.01, 2)
y= odeint(mmpa, yi, t, args= params)
```

<em>Exercise:</em> In the Michaelis-Menten approximation, the system simplifies to
$$
\begin{align}
\frac{dS}{dt} &= -\frac{k S E_\textrm{tot}}{\frac{b+k}{f} + S} \\
\frac{dP}{dt} &= \frac{k S E_\textrm{tot}}{\frac{b+k}{f} + S} 
\end{align}
$$
Write some code to simulate this system and compare (plot on the same graph) with the full system above for different ratios of initial <tt>S</tt> to total enzyme.

```python
def MM(y, t, params):
    f, b, k = params
    s, p, etot = y
    dydt = np.empty(len(y))
    dydt[0] = -(k*s*etot)/(((b+k)/f) + s)
    dydt[1] = k*s*etot/(((b+k)/f) + s)
    return dydt

params = np.array([1, 0.01, 4])
t= np.linspace(0, 800, 100)
yi= np.array([1000, 1, 1])
y = odeint(MM, yi, t, args = (params, ))
plt.figure()
plt.plot(t, y[:,0], t, y[:,1])
plt.show()
```
