```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def oocyte(y,t):
    f= 40
    n= 5
    K= 20
    k= 0.2
    p= 0
    dydt= np.empty(len(y))
    dydt= k*p + f*y**n/(K**n + y**n) - y
    return dydt
```

```python
init= [30]
t= np.arange(0,100, 0.1)
y= odeint(oocyte, init, t)

plt.figure()
plt.plot(t, y[:,0], label= 'active Mos')
plt.xlabel('time')
plt.legend()
plt.show()
```
