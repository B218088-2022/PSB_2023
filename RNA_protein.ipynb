```python
import numpy as np
import matplotlib.pylab as plt
from scipy.integrate import odeint
```

```python
u = 1
dm = np.log(2)/2
dp = np.log(2)/60
v = 0.8
```

```python
def gene(y, t):
    dydt = np.zeros(2)
    dydt[0] = u - dm *y[0]
    dydt[1] = v*y[0] - dp * y[1]
    return dydt
```

```python
t = np.linspace(0, 500, 100)
y = odeint(gene, [0,0], t)
```

```python
t
```

```python
plt.figure()
plt.plot(t, y[:, 0])
plt.plot(t, y[:,1])
plt.legend()
plt.show()
```
